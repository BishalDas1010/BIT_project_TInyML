#model Deployment




```python
import pandas as pd

# Features
X_train = pd.read_csv(
    "/content/extracted_files/UCI HAR Dataset/train/X_train.txt",
    delim_whitespace=True,
    header=None
)

X_test = pd.read_csv(
    "/content/extracted_files/UCI HAR Dataset/test/X_test.txt",
    delim_whitespace=True,
    header=None
)

# Labels
y_train = pd.read_csv(
    "/content/extracted_files/UCI HAR Dataset/train/y_train.txt",
    header=None
)

y_test = pd.read_csv(
    "/content/extracted_files/UCI HAR Dataset/test/y_test.txt",
    header=None
)
```

```python
(X_train.head())
```

```python
X_train.info()
```

```python
X_train.describe()
```

```python
X_train.shape
```

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(
    n_estimators=100,
    random_state=42,
    n_jobs=-1
)

rf.fit(X_train, y_train.values.ravel())
```

```python
y_pred = rf.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, classification_report

print("Accuracy:", accuracy_score(y_test, y_pred))

print(classification_report(y_test, y_pred))
```
