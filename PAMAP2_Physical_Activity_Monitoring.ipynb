```python
from sklearn.preprocessing import StandardScaler
import numpy as np

# The variables combine_df_x and Y must have the same indices for alignment
# Combine them temporarily to drop NaNs across both
combined_temp = pd.concat([combine_df_x, Y], axis=1).dropna()

# Separate back into X and Y
X_clean = combined_temp[feature].values
Y_clean = combined_temp['activityID']

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_clean)

print(f"Cleaned data shape: {X_scaled.shape}")
```

```python
X_windows = []
y_windows = []

window_size = 100
# Use X_scaled which was cleaned in wvh1CWx_kPym
limit = len(X_scaled)

for i in range(0, limit - window_size, window_size):
    X_windows.append(X_scaled[i:i+window_size])
    y_windows.append(Y_clean.iloc[i+window_size-1])

X_windows = np.array(X_windows, dtype='float32')
y_windows = np.array(y_windows)

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_windows, y_windows, test_size=0.2, random_state=42)

print("Data prepared. X_train shape:", X_train.shape)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X_windows,
    y_windows,
    test_size=0.2,
    random_state=42
)
```

```python
from sklearn.preprocessing import LabelEncoder
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv1D, MaxPooling1D, Flatten, Dense, Input

le = LabelEncoder()
y_train_mapped = le.fit_transform(y_train)
y_test_mapped = le.transform(y_test)

# Re-define model to ensure weight reset
model = Sequential([
    Input(shape=(100, 6)),
    Conv1D(filters=16, kernel_size=3, activation='relu'),
    MaxPooling1D(2),
    Conv1D(filters=32, kernel_size=3, activation='relu'),
    MaxPooling1D(2),
    Flatten(),
    Dense(32, activation='relu'),
    Dense(len(le.classes_), activation='softmax')
])

model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

history = model.fit(
    X_train,
    y_train_mapped,
    epochs=15,
    batch_size=64,
    validation_data=(X_test, y_test_mapped)
)
model.summary()

# Save the model after training
model.save("har_model.h5")
```

```python
# Use the mapped labels for evaluation to match the model's output layer
loss, acc = model.evaluate(X_test.astype('float32'), y_test_mapped)

print("Accuracy:", acc)
```
