```python
import pandas as pd
import numpy as np
from sklearn.ensemble import IsolationForest
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import seaborn as sns

model_df = df.copy()

# Handle missing values (critical for Isolation Forest)
print("Missing values before cleaning:")
print(model_df.isnull().sum())

# Option 1: Remove rows with any NaN (conservative)
# model_df = model_df.dropna()

# Option 2: Forward fill then backward fill (for time series)
model_df = model_df.fillna(method='ffill').fillna(method='bfill')

# Option 3: Interpolate (more sophisticated for time series)
# model_df = model_df.interpolate(method='linear')

print("\nMissing values after cleaning:")
print(model_df.isnull().sum())


# STEP 2: Feature Engineering (Optional but recommended)


# Add derived features that might help detect anomalies
model_df['HR_SBP_ratio'] = model_df['HR'] / model_df['SBP']
model_df['pulse_pressure'] = model_df['SBP'] - model_df['DBP']  # Normal: 30-50 mmHg
model_df['MAP'] = model_df['DBP'] + (model_df['pulse_pressure'] / 3)  # Mean Arterial Pressure

# Remove physiologically impossible values
model_df = model_df[
    (model_df['HR'].between(20, 250)) &
    (model_df['SBP'].between(50, 250)) &
    (model_df['DBP'].between(20, 150)) &
    (model_df['BT'].between(30, 42))  # Body temp in Celsius
]


# STEP 3: Select Features for Isolation Forest


# Basic features
feature_columns = ['HR', 'SBP', 'DBP', 'BT']

# Extended features (if you created derived features)
# feature_columns = ['HR', 'SBP', 'DBP', 'BT', 'HR_SBP_ratio', 'pulse_pressure', 'MAP']

X = model_df[feature_columns].values


# STEP 4: Normalize the data


scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)


# STEP 5: Apply Isolation Fores

# Initialize and train Isolation Forest
# contamination: expected proportion of anomalies (adjust based on your expectation)
iso_forest = IsolationForest(
    contamination=0.1,        # Expect ~10% anomalies (adjust as needed)
    n_estimators=100,          # Number of trees
    max_samples='auto',        # Number of samples per tree
    random_state=42,
    n_jobs=-1                  # Use all available cores
)

# Fit the model
iso_forest.fit(X_scaled)

# Get predictions: -1 for anomalies, 1 for normal
predictions = iso_forest.predict(X_scaled)

# Get anomaly scores (more negative = more anomalous)
anomaly_scores = iso_forest.decision_function(X_scaled)

# Add results to dataframe
model_df['anomaly'] = predictions
model_df['anomaly_score'] = anomaly_scores

# Convert to labels (0: normal, 1: anomaly)
model_df['anomaly_label'] = (model_df['anomaly'] == -1).astype(int)
```

```python


# STEP 6: Analyze Results

print("\n" + "="*50)
print("ISOLATION FOREST RESULTS")
print("="*50)

print(f"Total samples: {len(model_df)}")
print(f"Normal samples: {(model_df['anomaly_label'] == 0).sum()} ({(model_df['anomaly_label'] == 0).sum()/len(model_df)*100:.1f}%)")
print(f"Anomalies detected: {(model_df['anomaly_label'] == 1).sum()} ({(model_df['anomaly_label'] == 1).sum()/len(model_df)*100:.1f}%)")

# Statistics by anomaly status
print("\nFeature statistics by class:")
print(model_df.groupby('anomaly_label')[feature_columns].describe())


# STEP 7: Visualization


fig, axes = plt.subplots(3, 2, figsize=(15, 12))
fig.suptitle('Isolation Forest Anomaly Detection Results', fontsize=16)

# 1. Time series with anomalies highlighted
ax = axes[0, 0]
normal = model_df[model_df['anomaly_label'] == 0]
anomalies = model_df[model_df['anomaly_label'] == 1]
ax.scatter(normal['time'], normal['HR'], c='blue', alpha=0.5, s=1, label='Normal')
ax.scatter(anomalies['time'], anomalies['HR'], c='red', alpha=0.7, s=5, label='Anomaly')
ax.set_xlabel('Time (s)')
ax.set_ylabel('Heart Rate (bpm)')
ax.set_title('Heart Rate - Anomaly Detection')
ax.legend()
ax.grid(True, alpha=0.3)

# 2. SBP vs DBP scatter
ax = axes[0, 1]
scatter = ax.scatter(model_df['SBP'], model_df['DBP'],
                     c=model_df['anomaly_score'], cmap='coolwarm',
                     alpha=0.5, s=1)
ax.set_xlabel('SBP (mmHg)')
ax.set_ylabel('DBP (mmHg)')
ax.set_title('SBP vs DBP (colored by anomaly score)')
plt.colorbar(scatter, ax=ax, label='Anomaly Score')

# 3. Anomaly score distribution
ax = axes[1, 0]
ax.hist(model_df['anomaly_score'], bins=50, edgecolor='black', alpha=0.7)
ax.axvline(x=0, color='red', linestyle='--', label='Decision boundary')
ax.set_xlabel('Anomaly Score')
ax.set_ylabel('Frequency')
ax.set_title('Distribution of Anomaly Scores')
ax.legend()

# 4. Feature distributions by class
ax = axes[1, 1]
for i, feature in enumerate(feature_columns[:4]):
    normal_data = model_df[model_df['anomaly_label'] == 0][feature]
    anomaly_data = model_df[model_df['anomaly_label'] == 1][feature]

    positions = [i*2 - 0.3, i*2 + 0.3]
    bp = ax.boxplot([normal_data, anomaly_data], positions=positions,
                    widths=0.5, patch_artist=True)
    bp['boxes'][0].set_facecolor('lightblue')
    bp['boxes'][1].set_facecolor('lightcoral')

ax.set_xticks([i*2 for i in range(len(feature_columns))])
ax.set_xticklabels(feature_columns[:4])
ax.set_ylabel('Value')
ax.set_title('Feature Distributions: Normal vs Anomaly')
ax.legend([bp['boxes'][0], bp['boxes'][1]], ['Normal', 'Anomaly'])

# 5. Time-based anomaly pattern
ax = axes[2, 0]
time_bins = pd.cut(model_df['time'], bins=20)
anomaly_rate = model_df.groupby(time_bins)['anomaly_label'].mean()
ax.plot(range(len(anomaly_rate)), anomaly_rate.values, marker='o')
ax.set_xlabel('Time Bin')
ax.set_ylabel('Anomaly Rate')
ax.set_title('Anomaly Rate Over Time')
ax.grid(True, alpha=0.3)

# 6. Correlation between features and anomaly score
ax = axes[2, 1]
correlations = model_df[feature_columns + ['anomaly_score']].corr()['anomaly_score'].drop('anomaly_score')
correlations.plot(kind='barh', ax=ax, color='teal')
ax.set_xlabel('Correlation with Anomaly Score')
ax.set_title('Feature Importance for Anomaly Detection')
ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()


# STEP 8: Save Results


# Save labeled data
model_df.to_csv('vital_signs_with_anomaly_labels.csv', index=False)
print(f"\nResults saved to 'vital_signs_with_anomaly_labels.csv'")

# Save anomaly statistics
anomaly_stats = model_df.groupby('anomaly_label')[feature_columns].agg(['mean', 'std', 'min', 'max'])
anomaly_stats.to_csv('anomaly_statistics.csv')
print("Anomaly statistics saved to 'anomaly_statistics.csv'")
```

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv1D, MaxPooling1D, Flatten, Dense, Input, Dropout

# 1. PREPARE WINDOWED DATA (Your technique)


# Load your labeled data
Df = pd.read_csv("/content/vital_signs_with_anomaly_labels.csv")

# Select features and labels
features = ['HR', 'SBP', 'DBP', 'BT']  # 4 features
X_data = Df[features].values
y_data = Df['anomaly_label'].values  # 0 or 1

# Normalize
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_data)

print(X_scaled)
# Create windows (your sliding window technique)
X_windows = []
y_windows = []

window_size = 100  # 100 time steps

for i in range(0, len(X_scaled) - window_size, 10):  # Step=10 for overlap
    X_windows.append(X_scaled[i:i+window_size])
    # Use label from the LAST timestep in the window
    y_windows.append(y_data[i+window_size-1])

X_windows = np.array(X_windows, dtype='float32')
y_windows = np.array(y_windows, dtype='int')

print(f"X_windows shape: {X_windows.shape}")  # (samples, 100, 4)
print(f"y_windows shape: {y_windows.shape}")
print(f"Anomaly rate: {y_windows.mean()*100:.1f}%")


X_train, X_test, y_train, y_test = train_test_split(
    X_windows, y_windows, test_size=0.2, random_state=42, stratify=y_windows
)

print(f"\nX_train shape: {X_train.shape}")
print(f"X_test shape: {X_test.shape}")
print(f"Train anomaly rate: {y_train.mean()*100:.1f}%")
print(f"Test anomaly rate: {y_test.mean()*100:.1f}%")

# HANDLE CLASS IMBALANCE
# Option 1: Class weights
from sklearn.utils.class_weight import compute_class_weight

class_weights = compute_class_weight(
    'balanced',
    classes=np.array([0, 1]),
    y=y_train
)
class_weight_dict = {0: class_weights[0], 1: class_weights[1]}
print(f"\nClass weights: {class_weight_dict}")

```

```python

WINDOW_SIZE = 50  # Smaller window

# Recreate windows
X_windows_small = []
y_windows_small = []

for i in range(0, len(X_scaled) - WINDOW_SIZE, 5):
    X_windows_small.append(X_scaled[i:i+WINDOW_SIZE])
    y_windows_small.append(y_data[i+WINDOW_SIZE-1])

X_windows_small = np.array(X_windows_small, dtype='float32')
y_windows_small = np.array(y_windows_small, dtype='int')

X_train_s, X_test_s, y_train_s, y_test_s = train_test_split(
    X_windows_small, y_windows_small, test_size=0.2, random_state=42
)

# Ultra-compact model
tiny_model = Sequential([
    Input(shape=(WINDOW_SIZE, 4)),

    # Single conv layer
    Conv1D(filters=8, kernel_size=5, activation='relu', padding='same'),
    MaxPooling1D(4),  # Output: (12, 8)

    Conv1D(filters=16, kernel_size=3, activation='relu', padding='same'),
    MaxPooling1D(2),  # Output: (6, 16)

    Flatten(),

    Dense(8, activation='relu'),
    Dense(1, activation='sigmoid')
])

tiny_model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)

tiny_model.summary()  # Much fewer parameters!

# Train
history_tiny = tiny_model.fit(
    X_train_s, y_train_s,
    epochs=15,
    batch_size=32,
    validation_data=(X_test_s, y_test_s),
    class_weight=class_weight_dict
)

# Convert to TFLite for ESP32
converter = tf.lite.TFLiteConverter.from_keras_model(tiny_model)
converter.optimizations = [tf.lite.Optimize.DEFAULT]
tflite_model = converter.convert()

with open('anomaly_detector_esp32.tflite', 'wb') as f:
    f.write(tflite_model)

print(f"TFLite model size: {len(tflite_model)} bytes")
```

```python
def predict_vitals(model, scaler, window_data, threshold=0.25):
    """Predict if vital signs are anomalous"""
    window_scaled = scaler.transform(window_data)
    window_input = window_scaled.reshape(1, window_data.shape[0], window_data.shape[1])
    prob = model.predict(window_input, verbose=1)[0][0]
    pred = 1 if prob > threshold else 0
    return pred, prob

# Test it
for i in range(10):
    test_window = X_test_s[i]
    pred, prob = predict_vitals(tiny_model, scaler, test_window)
    actual = y_test_s[i]

    status = "ANOMALY" if pred == 1 else "✓ Normal"
    actual_status = "ANOMALY" if actual == 1 else "Normal"
    match = "✓" if pred == actual else "✗"

    print(f"{match} Sample {i}: {status} (conf: {prob:.3f}) | Actual: {actual_status}")
```
